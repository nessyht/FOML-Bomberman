# Linear Regression with sklearn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set() # set seaborn as plotting style
```

## Simple linear fit

```python
# create function of form y = a*x + b  with a=2; b=-5
# with scattered y-values

rng = np.random.RandomState(1)
x = 10*rng.rand(50) # rand(): uniform distr
y = 2*x-5+rng.randn(50) # randn(): normal distr
fig, ax = plt.subplots(1,1,figsize=(12,6),dpi=200)
plt.scatter(x,y)
plt.show()
```

Scikit-Learn clearly distinguishes between the <i>choice of a model</i> and the <i> application of the model to the data</i>:

```python
from sklearn.linear_model import LinearRegression

# Choose model and hyperparameters
model = LinearRegression(fit_intercept=True)

# Apply model to data (training)
model.fit(x[:,np.newaxis], y)   
'''
   model.fit(X,y) expects X to a 2D feature matrix of shape (n_instances, n_features).
   Therefore we have to transform x into an array of that shape.
'''
xfit = np.linspace(0,10,1000)

# Predict with trained model
yfit = model.predict(xfit[:,np.newaxis])

# plotting results
fig, ax = plt.subplots(1,1,figsize=(12,6),dpi=200)
plt.scatter(x,y)
plt.plot(xfit, yfit, '-', lw=1, color='red', alpha=0.5)
plt.show()
```

```python
print('a = ', model.coef_[0])
print('b = ', model.intercept_)
```

## Example with feature matrix and target vector

```python
# Example with 100 instances with 3 features each

X = 10 * rng.rand(100,3)
y = 0.5 + np.dot(X, [1.5, -2., 1.]) # targets

# Fit will be exact since we didn't include any scattering
model.fit(X,y)

print(model.intercept_)
print(model.coef_)
```
