# Principle Component Analysis (PCA)
## A simple example

```python
import seaborn as sns
sns.set()

iris = sns.load_dataset('iris')
iris.head()
```

```python
# Create feature matrix and target vector
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']
```

```python
from sklearn.decomposition import PCA

# Choose model to reduce dimensions to 2; Before was 4 (number of features)
model = PCA(n_components=2) 

# Fit model to data
model.fit(X_iris) # No target vector! Because PCA is a transformation, no classification.

# Transform data into 2D
X_2D = model.transform(X_iris)
```

```python
# Add new columns to data frame
iris['PCA1'] = X_2D[:,0]
iris['PCA2'] = X_2D[:,1]

# Plot
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False, size=10)
```
